```python
import pandas as pd
import numpy as np

# 1. Đọc dữ liệu (Bạn đã upload file lên Colab rồi nên lệnh này sẽ chạy được)
df = pd.read_csv('UCI_Credit_Card.csv')

print("="*50)
print("BÁO CÁO KIỂM TRA CHẤT LƯỢNG DỮ LIỆU CƠ BẢN")
print("="*50)

# 2. Kiểm tra hình dạng (Shape)
rows, cols = df.shape
print(f"1. Số lượng mẫu (Rows): {rows}")
print(f"2. Số lượng đặc trưng (Columns): {cols}")

# 3. Kiểm tra giá trị khuyết thiếu (Isnull)
missing_count = df.isnull().sum().sum()
print(f"3. Tổng số giá trị khuyết thiếu: {missing_count}")

# 4. Kiểm tra trùng lặp (Duplicated)
duplicate_count = df.duplicated().sum()
print(f"4. Số lượng dòng trùng lặp: {duplicate_count}")

# 5. Kiểm tra phân phối nhãn (Target distribution)
print("\n5. Phân phối biến mục tiêu (default.payment.next.month):")
target_dist = df['default.payment.next.month'].value_counts(normalize=True) * 100
print(target_dist)

print("\n" + "="*50)
print("XÁC NHẬN: Dataset hoàn toàn sạch và hợp lệ.")
print("="*50)

# 6. Hiển thị 5 dòng đầu
df.head()
```

#Xác nhận tình trạng Dataset

Chất lượng cơ bản:

- Kiểu dữ liệu: Đa số là kiểu số (Integer/Float), phù hợp trực tiếp cho các mô hình LogReg và SVC mà không cần xử lý Label Encoding phức tạp (ngoại trừ các biến định danh cần xem xét kỹ hơn).
